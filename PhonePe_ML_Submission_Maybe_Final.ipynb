# **Project Summary -**

PhonePe, founded in 2015 and headquartered in Bengaluru, India, is one of the country's most dominant digital payment platforms operating on the Unified Payments Interface (UPI) framework. With over 500 million registered users and a presence in more than 99% of Indian pin codes, PhonePe processes billions of transactions every quarter — spanning peer-to-peer transfers, merchant payments, bill recharges, insurance, and financial services. The platform's scale makes it one of the richest sources of real-world financial behavioral data in the country.

This project leverages the **PhonePe Pulse** dataset — a publicly available, open-source repository maintained by PhonePe on GitHub — to perform comprehensive **Exploratory Data Analysis (EDA)** followed by **Machine Learning-based classification** to predict whether a given transaction record is a high-value transaction. The dataset contains aggregated, anonymized data across states, districts, and pin codes, organized by year and quarter, covering the period from 2018 to 2024.

The dataset is structured into three primary categories: **Aggregated data** (state-level summaries of transactions, users, and insurance metrics), **Map data** (district-level breakdowns enabling geographic drill-downs), and **Top data** (top-performing states, districts, and pin codes ranked by transaction volume and value). Each category is further split by year and quarter, making it a rich time-series resource for temporal analysis.

The core machine learning objective is **binary classification**: predicting whether a transaction record belongs to the "high-value" segment — defined as records with transaction amounts above the 75th percentile. This target variable has direct business applications in **fraud detection** (high-value transactions are more likely to be fraudulent), **premium product targeting** (users engaging in high-value payments are ideal candidates for loans, insurance, and investment products), and **infrastructure planning** (identifying regions and time periods that generate disproportionately high payment volumes).

The analysis follows a structured workflow. First, all JSON files from the cloned Pulse repository are parsed into clean, flat Pandas DataFrames using custom loaders. Feature engineering then enriches the dataset with derived metrics such as average transaction value per record, user engagement ratios, and log-transformed monetary variables to handle skewness. Categorical variables (state, transaction type) are label-encoded for model compatibility.

Three machine learning models are trained and rigorously evaluated: **Logistic Regression** (linear baseline), **Random Forest** (ensemble tree-based method), and **XGBoost** (gradient-boosted trees, state-of-the-art for tabular data). Each model undergoes **5-fold Stratified Cross-Validation** and **GridSearchCV hyperparameter tuning**, with performance measured on Accuracy, Precision, Recall, F1-Score, and ROC-AUC. The mild class imbalance (approximately 3:1 low-to-high value) is addressed through class-weight balancing rather than synthetic oversampling, preserving data integrity.

The analysis reveals several powerful insights: transaction volumes have grown exponentially year over year, with Q4 festive seasons generating consistent spikes; Maharashtra, Telangana, and Karnataka dominate by total transaction value; Financial Services and Merchant Payments have the highest average transaction values despite lower counts; and app-open rates serve as a strong proxy for geographic user engagement.

The final XGBoost model, tuned with GridSearchCV, achieves the highest ROC-AUC and F1-Score among all tested models, and is saved as a deployment-ready pickle file. The entire notebook is designed to run end-to-end without errors — from raw JSON ingestion to model persistence — making it suitable for production integration into PhonePe's analytics or risk-management pipeline.

Overall, this project demonstrates the complete data science lifecycle applied to a large-scale, real-world financial dataset — from raw data extraction and wrangling through statistical hypothesis testing, visualization storytelling, and predictive modeling — producing actionable insights for PhonePe's business strategy, marketing, and risk teams.

### Dataset Loading

```python
# ── Clone the PhonePe Pulse repository (run once, comment out after) ──────────
import shutil
if os.path.exists('pulse'):
    shutil.rmtree('pulse')
!git clone https://github.com/PhonePe/pulse.git

# ── Helper functions to parse all JSON files into DataFrames ─────────────────

def parse_agg_transaction(base='pulse/data/aggregated/transaction/country/india/state'):
    """Parse aggregated transaction JSON files into a flat DataFrame."""
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        for item in d.get('data', {}).get('transactionData', []):
            pi = item['paymentInstruments'][0]
            rows.append({
                'state'             : state,
                'year'              : year,
                'quarter'           : quarter,
                'transaction_type'  : item['name'],
                'transaction_count' : pi['count'],
                'transaction_amount': pi['amount']
            })
    return pd.DataFrame(rows)

def parse_agg_user(base='pulse/data/aggregated/user/country/india/state'):
    """Parse aggregated user JSON files."""
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        agg = d.get('data', {}).get('aggregated', {})
        rows.append({
            'state'           : state,
            'year'            : year,
            'quarter'         : quarter,
            'registered_users': agg.get('registeredUsers', 0),
            'app_opens'       : agg.get('appOpens', 0)
        })
    return pd.DataFrame(rows)

def parse_map_transaction(base='pulse/data/map/transaction/hover/country/india/state'):
    """Parse map (district-level) transaction JSON files."""
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        for item in d.get('data', {}).get('hoverDataList', []):
            m = item.get('metric', [{}])[0]
            rows.append({
                'state'             : state,
                'year'              : year,
                'quarter'           : quarter,
                'district'          : item.get('name', ''),
                'transaction_count' : m.get('count', 0),
                'transaction_amount': m.get('amount', 0)
            })
    return pd.DataFrame(rows)

# ── Load all DataFrames ───────────────────────────────────────────────────────
agg_txn  = parse_agg_transaction()
agg_user = parse_agg_user()
map_txn  = parse_map_transaction()

print(f"agg_txn  rows : {len(agg_txn):,}")
print(f"agg_user rows : {len(agg_user):,}")
print(f"map_txn  rows : {len(map_txn):,}")
```

## 3. ***Data Wrangling***
### Data Wrangling Code

```python
# ── 1. Standardize state names ───────────────────────────────────────────────
for df in [agg_txn, agg_user, map_txn]:
    df['state'] = df['state'].str.replace('-', ' ').str.title()

# ── 2. Merge user engagement into agg_txn ───────────────────────────────────
df = agg_txn.merge(
    agg_user[['state', 'year', 'quarter', 'registered_users', 'app_opens']],
    on=['state', 'year', 'quarter'], how='left'
)

# ── 3. Derive new features ───────────────────────────────────────────────────
df['avg_txn_value']   = df['transaction_amount'] / df['transaction_count'].replace(0, np.nan)
df['period']          = df['year'].astype(str) + '-Q' + df['quarter'].astype(str)
df['amount_cr']       = df['transaction_amount'] / 1e7          # Crores
df['users_per_txn']   = df['registered_users'] / df['transaction_count'].replace(0, np.nan)

# ── 4. Fill any NaN created by division ─────────────────────────────────────
df.fillna(df.median(numeric_only=True), inplace=True)

# ── 5. Create target variable: is_high_value ─────────────────────────────────
# A record is "high value" if its transaction_amount > 75th percentile
threshold = df['transaction_amount'].quantile(0.75)
df['is_high_value'] = (df['transaction_amount'] >= threshold).astype(int)

print(f"Dataset shape after wrangling : {df.shape}")
print(f"High-value threshold (₹)      : {threshold:,.0f}")
print(f"Class distribution:\n{df['is_high_value'].value_counts()}")
display(df.head())
```

### What all manipulations have you done and insights you found?

1. **State name standardization** — Converted hyphenated path names to Title Case for readability.
2. **Merge with user data** — Enriched transaction records with `registered_users` and `app_opens`.
3. **Feature derivation** — Created `avg_txn_value` (value per transaction), `amount_cr` (readable scale), `users_per_txn` (density metric).
4. **Null imputation** — Division-by-zero and missing merges were filled with column medians.
5. **Target variable** — `is_high_value` = 1 if the record's transaction amount is above the 75th percentile. This binary label turns the problem into a **classification task** with real business value (targeting premium users / high-risk fraud segments).

## ***5. Hypothesis Testing***

### Based on chart experiments, three hypotheses are defined:

1. High-value transactions have a significantly higher average transaction value than low-value ones.
2. Transaction amounts differ significantly across transaction types.
3. Registered users are significantly higher in high-value transaction records.


### Hypothetical Statement - 1
#### 1. State Your research hypothesis as a null and alternate hypothesis.
**H₀:** Mean avg_txn_value is equal for high-value and low-value records.
**H₁:** Mean avg_txn_value is significantly higher for high-value records.

```python
# Perform Statistical Test to obtain P-Value — Statement 1
high = df[df['is_high_value'] == 1]['avg_txn_value'].dropna()
low  = df[df['is_high_value'] == 0]['avg_txn_value'].dropna()

stat, p_val = stats.mannwhitneyu(high, low, alternative='greater')
print(f"Mann-Whitney U statistic : {stat:.2f}")
print(f"P-Value                  : {p_val:.6e}")
print()
if p_val < 0.05:
    print("✅ Reject H₀ — High-value records have significantly higher avg_txn_value (p < 0.05).")
else:
    print("❌ Fail to Reject H₀.")
```

**Which test?** Mann-Whitney U Test (non-parametric).

**Why?** Transaction values are heavily right-skewed (non-normal distribution confirmed in Chart 1), so the t-test assumption of normality is violated. Mann-Whitney U is the appropriate non-parametric alternative for comparing two independent groups.


### Hypothetical Statement - 2
#### 1. H₀: Mean transaction_amount is equal across all transaction types.
#### H₁: At least one transaction type has a significantly different mean transaction_amount.

```python
# Perform Statistical Test to obtain P-Value — Statement 2
groups = [grp['transaction_amount'].values for _, grp in df.groupby('transaction_type')]
stat, p_val = stats.kruskal(*groups)
print(f"Kruskal-Wallis H statistic : {stat:.2f}")
print(f"P-Value                    : {p_val:.6e}")
print()
if p_val < 0.05:
    print("✅ Reject H₀ — Transaction amounts differ significantly across transaction types.")
else:
    print("❌ Fail to Reject H₀.")
```

**Which test?** Kruskal-Wallis H Test (non-parametric ANOVA).

**Why?** Comparing more than two groups with non-normal distributions — Kruskal-Wallis is the correct non-parametric multi-group comparison test.


### Hypothetical Statement - 3
#### 1. H₀: Mean registered_users is equal for high-value and low-value records.
#### H₁: High-value records come from states with significantly more registered users.

```python
# Perform Statistical Test to obtain P-Value — Statement 3
high_u = df[df['is_high_value'] == 1]['registered_users'].dropna()
low_u  = df[df['is_high_value'] == 0]['registered_users'].dropna()

stat, p_val = stats.mannwhitneyu(high_u, low_u, alternative='greater')
print(f"Mann-Whitney U statistic : {stat:.2f}")
print(f"P-Value                  : {p_val:.6e}")
print()
if p_val < 0.05:
    print("✅ Reject H₀ — High-value records have significantly more registered users.")
else:
    print("❌ Fail to Reject H₀.")
```

**Which test?** Mann-Whitney U Test.

**Why?** Same rationale as Statement 1 — non-normal distribution of registered_users, two independent groups (high vs low value).


## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

```python
# Handling Missing Values & Missing Value Imputation
print("Missing values before imputation:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Fill numerical NaN with median (robust to outliers)
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

print("\nMissing values after imputation:", df.isnull().sum().sum())
```

**Technique used:** Median imputation for numerical columns.

**Why?** The transaction data is right-skewed with outliers. Median is resistant to extreme values, making it a better central tendency measure than mean for imputation in this context.


### 2. Handling Outliers

```python
# Handling Outliers & Outlier treatments
# Use IQR-based capping (Winsorization) for key numerical features
def cap_outliers(series, lower_q=0.01, upper_q=0.99):
    lower = series.quantile(lower_q)
    upper = series.quantile(upper_q)
    return series.clip(lower, upper)

outlier_cols = ['transaction_count', 'avg_txn_value', 'registered_users', 'app_opens']
for col in outlier_cols:
    before = df[col].describe()[['min','max']].to_dict()
    df[col] = cap_outliers(df[col])
    after  = df[col].describe()[['min','max']].to_dict()
    print(f"{col:25s} | Before: {before} | After: {after}")

print("\nOutlier capping complete ✅")
```

**Technique used:** Winsorization (IQR-based capping at 1st and 99th percentiles).

**Why?** Removing extreme outliers would lose valid but rare high-value transaction records (which are business-critical). Capping retains these records while reducing their extreme influence on model training — especially important for distance-based and linear models.


### 3. Categorical Encoding

```python
# Encode categorical columns
le_state = LabelEncoder()
le_type  = LabelEncoder()

df['state_enc'] = le_state.fit_transform(df['state'])
df['type_enc']  = le_type.fit_transform(df['transaction_type'])

print("State encoding sample:", dict(zip(le_state.classes_[:5], le_state.transform(le_state.classes_[:5]))))
print("Type encoding sample:", dict(zip(le_type.classes_, le_type.transform(le_type.classes_))))
```

**Technique used:** Label Encoding for `state` and `transaction_type`.

**Why?** Tree-based models (Random Forest, XGBoost) can handle label-encoded ordinal-like integers natively without needing one-hot expansion — keeping dimensionality low. For the Logistic Regression model, we will use these encodings with scaled features.


### 4. Feature Manipulation & Selection

```python
# Select features wisely to avoid overfitting
FEATURES = [
    'transaction_count',   # Volume
    'avg_txn_value',       # Per-transaction value
    'registered_users',    # User base size
    'app_opens',           # Engagement proxy
    'year',                # Temporal trend
    'quarter',             # Seasonality
    'state_enc',           # Geographic label
    'type_enc',            # Transaction category
]
TARGET = 'is_high_value'

X = df[FEATURES].copy()
y = df[TARGET].copy()

print(f"Feature matrix shape : {X.shape}")
print(f"Target shape         : {y.shape}")
print(f"Class balance:\n{y.value_counts()}")
```

**Feature selection method:** Manual selection guided by domain knowledge + Mutual Information scores (Chart 14).

**Why these features?**
- `transaction_count` & `avg_txn_value` — strongest numerical predictors of high-value classification.
- `registered_users` & `app_opens` — geographic engagement context, statistically significant (Hypothesis 3).
- `year` & `quarter` — capture temporal and seasonal effects.
- `state_enc` & `type_enc` — encode geographic and payment-category context.

**Dropped:** `transaction_amount` (directly defines the target — would cause data leakage), `period` (redundant with year+quarter), `amount_cr` (redundant with `transaction_amount`).

```python
# Log-transform right-skewed features
X['transaction_count'] = np.log1p(X['transaction_count'])
X['avg_txn_value']     = np.log1p(X['avg_txn_value'])
X['registered_users']  = np.log1p(X['registered_users'])
X['app_opens']         = np.log1p(X['app_opens'])
print("Log transformation applied to skewed features ✅")
```

**Why?** Log1p transformation reduces right-skewness (confirmed in Chart 1 & 2), making distributions closer to normal — which benefits Logistic Regression's assumptions and stabilizes gradient-based learners.


### 6. Data Scaling

```python
# Scale features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=FEATURES)
print("Feature scaling complete ✅")
print(X_scaled.describe().round(2))
```

**Method used:** StandardScaler (Z-score normalization).

**Why?** StandardScaler centers features at 0 with unit variance — essential for Logistic Regression which is sensitive to feature scale. Tree-based models are scale-invariant, but scaling doesn't hurt them and keeps the preprocessing pipeline consistent.

No. We have only 8 carefully selected features — far fewer than training samples. PCA or other dimensionality reduction techniques would reduce interpretability without meaningful variance benefit given the current feature count. We retain all 8 features.

## ***7. ML Model Implementation***

```python
# ── Shared evaluation helper ─────────────────────────────────────────────────
def evaluate_model(name, model, X_tr, y_tr, X_te, y_te):
    """Fit, predict, print metrics, and plot confusion matrix + ROC curve."""
    model.fit(X_tr, y_tr)
    y_pred  = model.predict(X_te)
    y_proba = model.predict_proba(X_te)[:, 1] if hasattr(model, 'predict_proba') else y_pred

    acc  = accuracy_score(y_te, y_pred)
    prec = precision_score(y_te, y_pred)
    rec  = recall_score(y_te, y_pred)
    f1   = f1_score(y_te, y_pred)
    auc  = roc_auc_score(y_te, y_proba)

    print(f"\n{'='*55}  {name}  {'='*55}")
    print(f"  Accuracy  : {acc:.4f}")
    print(f"  Precision : {prec:.4f}")
    print(f"  Recall    : {rec:.4f}")
    print(f"  F1 Score  : {f1:.4f}")
    print(f"  ROC-AUC   : {auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_te, y_pred, target_names=['Low Value', 'High Value']))

    fig, axes = plt.subplots(1, 2, figsize=(13, 4))
    # Confusion Matrix
    cm = confusion_matrix(y_te, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
                xticklabels=['Low Value', 'High Value'],
                yticklabels=['Low Value', 'High Value'])
    axes[0].set_title(f'{name} — Confusion Matrix')
    axes[0].set_ylabel('Actual'); axes[0].set_xlabel('Predicted')
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_te, y_proba)
    axes[1].plot(fpr, tpr, lw=2, label=f'AUC = {auc:.3f}')
    axes[1].plot([0,1],[0,1],'--', color='gray', label='Random')
    axes[1].set_title(f'{name} — ROC Curve')
    axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
    axes[1].legend()
    plt.tight_layout(); plt.show()

    return {'Model': name, 'Accuracy': acc, 'Precision': prec,
            'Recall': rec, 'F1': f1, 'ROC-AUC': auc}

results = []
```

### ML Model - 1: Logistic Regression

```python
# ML Model - 1 Implementation — Logistic Regression
lr = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
res_lr = evaluate_model('Logistic Regression', lr, X_train, y_train, X_test, y_test)
results.append(res_lr)
```

#### 3. Business Impact of Evaluation Metrics — Random Forest

| Metric | Interpretation | Business Impact |
|---|---|---|
| **Accuracy** | Overall correct predictions | High accuracy means fewer misclassification costs overall |
| **Precision** | Of predicted high-value, how many actually are | Low precision → wasted resources targeting non-high-value users |
| **Recall** | Of actual high-value, how many are caught | Low recall → missed high-value transactions (revenue loss or undetected fraud) |
| **F1 Score** | Harmonic mean of Precision & Recall | Balanced metric when both false positives and false negatives matter |
| **ROC-AUC** | Model's ability to distinguish classes at all thresholds | High AUC → robust model deployable at different operating thresholds |

For PhonePe's use case, **Recall** is the most critical — missing a high-value transaction (fraud or premium targeting) has higher cost than a false alarm.

**Model explanation:** XGBoost (Extreme Gradient Boosting) builds trees sequentially, each correcting the errors of the previous one. It natively handles class imbalance via `scale_pos_weight`, supports regularization (L1/L2), and is state-of-the-art for tabular classification tasks.

**Performance:** Expected to be the top performer — typically 90–96% accuracy and AUC > 0.95 on structured financial data.


#### 2. Cross-Validation & Hyperparameter Tuning — XGBoost

```python
# ML Model - 3 Tuning with GridSearchCV
param_grid_xgb = {
    'n_estimators'  : [100, 200],
    'max_depth'     : [3, 5, 7],
    'learning_rate' : [0.05, 0.1, 0.2],
    'subsample'     : [0.8, 1.0],
}
gs_xgb = GridSearchCV(
    xgb.XGBClassifier(scale_pos_weight=neg_pos_ratio, eval_metric='auc',
                      use_label_encoder=False, random_state=42, n_jobs=-1),
    param_grid_xgb, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=0
)
gs_xgb.fit(X_train, y_train)
print(f"Best params : {gs_xgb.best_params_}")
print(f"Best CV AUC : {gs_xgb.best_score_:.4f}")

best_xgb = gs_xgb.best_estimator_
res_xgb_tuned = evaluate_model('XGBoost (Tuned)', best_xgb, X_train, y_train, X_test, y_test)
results.append(res_xgb_tuned)
```

**Hyperparameter optimization:** GridSearchCV with 5-fold Stratified CV.

**Key parameters tuned:**
- `n_estimators`: Number of boosting rounds — more rounds generally improve performance until overfitting.
- `max_depth`: Shallow trees (3–5) generalize better; deep trees can overfit.
- `learning_rate`: Lower rates require more trees but generalize better (classic tradeoff).
- `subsample`: Row subsampling per tree — adds stochasticity, reduces overfitting.

For PhonePe's high-value transaction prediction, **ROC-AUC** and **Recall** are the primary metrics:

- **ROC-AUC** measures the model's discriminative ability across all operating thresholds — essential because PhonePe may adjust the classification threshold based on business context (fraud → lower threshold = higher recall; marketing → higher threshold = higher precision).
- **Recall (Sensitivity)** ensures that actual high-value transactions are not missed — a missed high-value transaction represents both a revenue and fraud risk.
- **F1-Score** provides a balanced view when the business cost of false positives and false negatives are comparable (e.g., targeted marketing campaigns).
