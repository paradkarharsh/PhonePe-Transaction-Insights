# **Project Summary -**

PhonePe, founded in 2015 and headquartered in Bengaluru, India, is one of the country's most dominant digital payment platforms operating on the Unified Payments Interface (UPI) framework. With over 500 million registered users and a presence in more than 99% of Indian pin codes, PhonePe processes billions of transactions every quarter — spanning peer-to-peer transfers, merchant payments, bill recharges, insurance, and financial services. The platform's scale makes it one of the richest sources of real-world financial behavioral data in the country.

This project leverages the **PhonePe Pulse** dataset — a publicly available, open-source repository maintained by PhonePe on GitHub — to perform comprehensive **Exploratory Data Analysis (EDA)** followed by **Machine Learning-based classification** to predict whether a given transaction record is a high-value transaction. The dataset contains aggregated, anonymized data across states, districts, and pin codes, organized by year and quarter, covering the period from 2018 to 2024.

The dataset is structured into three primary categories: **Aggregated data** (state-level summaries of transactions, users, and insurance metrics), **Map data** (district-level breakdowns enabling geographic drill-downs), and **Top data** (top-performing states, districts, and pin codes ranked by transaction volume and value). Each category is further split by year and quarter, making it a rich time-series resource for temporal analysis.

The core machine learning objective is **binary classification**: predicting whether a transaction record belongs to the "high-value" segment — defined as records with transaction amounts above the 75th percentile. This target variable has direct business applications in **fraud detection** (high-value transactions are more likely to be fraudulent), **premium product targeting** (users engaging in high-value payments are ideal candidates for loans, insurance, and investment products), and **infrastructure planning** (identifying regions and time periods that generate disproportionately high payment volumes).

The analysis follows a structured workflow. First, all JSON files from the cloned Pulse repository are parsed into clean, flat Pandas DataFrames using custom loaders. Feature engineering then enriches the dataset with derived metrics such as average transaction value per record, user engagement ratios, and log-transformed monetary variables to handle skewness. Categorical variables (state, transaction type) are label-encoded for model compatibility.

Three machine learning models are trained and rigorously evaluated: **Logistic Regression** (linear baseline), **Random Forest** (ensemble tree-based method), and **XGBoost** (gradient-boosted trees, state-of-the-art for tabular data). Each model undergoes **5-fold Stratified Cross-Validation** and **GridSearchCV hyperparameter tuning**, with performance measured on Accuracy, Precision, Recall, F1-Score, and ROC-AUC. The mild class imbalance (approximately 3:1 low-to-high value) is addressed through class-weight balancing rather than synthetic oversampling, preserving data integrity.

The analysis reveals several powerful insights: transaction volumes have grown exponentially year over year, with Q4 festive seasons generating consistent spikes; Maharashtra, Telangana, and Karnataka dominate by total transaction value; Financial Services and Merchant Payments have the highest average transaction values despite lower counts; and app-open rates serve as a strong proxy for geographic user engagement.

The final XGBoost model, tuned with GridSearchCV, achieves the highest ROC-AUC and F1-Score among all tested models, and is saved as a deployment-ready pickle file. The entire notebook is designed to run end-to-end without errors — from raw JSON ingestion to model persistence — making it suitable for production integration into PhonePe's analytics or risk-management pipeline.

### Dataset Loading

````python
# Load Dataset
# ── Clone the PhonePe Pulse repository (run once, then comment out) ───────────
!git clone https://github.com/PhonePe/pulse.git

# ── Parser: Aggregated Transaction ───────────────────────────────────────────
def parse_agg_transaction(base='pulse/data/aggregated/transaction/country/india/state'):
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        for item in d.get('data', {}).get('transactionData', []):
            pi = item['paymentInstruments'][0]
            rows.append({'state': state, 'year': year, 'quarter': quarter,
                         'transaction_type': item['name'],
                         'transaction_count': pi['count'],
                         'transaction_amount': pi['amount']})
    return pd.DataFrame(rows)

# ── Parser: Aggregated User ───────────────────────────────────────────────────
def parse_agg_user(base='pulse/data/aggregated/user/country/india/state'):
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        agg = d.get('data', {}).get('aggregated', {})
        rows.append({'state': state, 'year': year, 'quarter': quarter,
                     'registered_users': agg.get('registeredUsers', 0),
                     'app_opens': agg.get('appOpens', 0)})
    return pd.DataFrame(rows)

# ── Parser: Map Transaction ───────────────────────────────────────────────────
def parse_map_transaction(base='pulse/data/map/transaction/hover/country/india/state'):
    rows = []
    for fp in glob.glob(f'{base}/**/*.json', recursive=True):
        parts = fp.replace('\\', '/').split('/')
        state   = parts[-3]
        year    = int(parts[-2])
        quarter = int(parts[-1].replace('.json', ''))
        with open(fp) as f:
            d = json.load(f)
        for item in d.get('data', {}).get('hoverDataList', []):
            m = item.get('metric', [{}])[0]
            rows.append({'state': state, 'year': year, 'quarter': quarter,
                         'district': item.get('name', ''),
                         'transaction_count': m.get('count', 0),
                         'transaction_amount': m.get('amount', 0)})
    return pd.DataFrame(rows)

# ── Load all DataFrames ───────────────────────────────────────────────────────
agg_txn  = parse_agg_transaction()
agg_user = parse_agg_user()
map_txn  = parse_map_transaction()

print(f"agg_txn  rows : {len(agg_txn):,}")
print(f"agg_user rows : {len(agg_user):,}")
print(f"map_txn  rows : {len(map_txn):,}")
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Standardize state names — replace hyphens with spaces, apply Title Case
for df in [agg_txn, agg_user, map_txn]:
    df['state'] = df['state'].str.replace('-', ' ').str.title()

# 2. Merge user engagement features into the main transaction DataFrame
df = agg_txn.merge(
    agg_user[['state', 'year', 'quarter', 'registered_users', 'app_opens']],
    on=['state', 'year', 'quarter'], how='left'
)

# 3. Derive new features
df['avg_txn_value'] = df['transaction_amount'] / df['transaction_count'].replace(0, np.nan)
df['amount_cr']     = df['transaction_amount'] / 1e7          # Convert to Crores for readability
df['period']        = df['year'].astype(str) + '-Q' + df['quarter'].astype(str)
df['users_per_txn'] = df['registered_users'] / df['transaction_count'].replace(0, np.nan)

# 4. Fill NaN values created by division with column medians
df.fillna(df.median(numeric_only=True), inplace=True)

# 5. Create the target variable: is_high_value
# A record is "high value" if transaction_amount >= 75th percentile
threshold = df['transaction_amount'].quantile(0.75)
df['is_high_value'] = (df['transaction_amount'] >= threshold).astype(int)

print(f"Final DataFrame shape     : {df.shape}")
print(f"High-value threshold (₹)  : {threshold:,.0f}")
print(f"\nClass distribution:")
print(df['is_high_value'].value_counts())
display(df.head())
````

### What all manipulations have you done and insights you found?

1. **State name standardization** — Raw JSON path names used hyphens (e.g., `andhra-pradesh`). Converted to readable Title Case (e.g., `Andhra Pradesh`) for consistent merging and display.

2. **User data merge** — Enriched the main `agg_txn` DataFrame with `registered_users` and `app_opens` from `agg_user` by joining on state + year + quarter. This added two strong engagement features.

3. **Feature derivation**:
   - `avg_txn_value` — average rupee value per individual transaction (amount ÷ count). Strong predictor of high-value classification.
   - `amount_cr` — transaction amount in Crores for human-readable axis labels in charts.
   - `period` — string label combining year and quarter (e.g., `2022-Q3`) for time-series x-axis.
   - `users_per_txn` — ratio of registered users to transactions, a density/saturation metric.

4. **Null imputation** — Division by zero and unmatched merge rows produced NaN. Filled with column medians (robust to skewness).

5. **Target variable creation** — `is_high_value` = 1 if `transaction_amount` ≥ 75th percentile, else 0. This converts the problem into a binary classification task directly relevant to fraud detection and premium product targeting.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Based on the chart analysis, the following three hypotheses are defined:

1. **High-value transaction records have a significantly higher average transaction value (`avg_txn_value`) than low-value records.** (Observed in Chart 12 — class KDE distributions on `avg_txn_value` are clearly separated.)

2. **Transaction amounts differ significantly across different transaction types.** (Observed in Chart 6 — box plots show clearly different medians across categories.)

3. **High-value records are associated with significantly higher registered user counts than low-value records.** (Observed in Chart 7 and Chart 15 — high-value points cluster in higher `registered_users` regions.)

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):** The mean `avg_txn_value` of high-value records (is_high_value = 1) is **equal to** the mean `avg_txn_value` of low-value records (is_high_value = 0).

**Alternate Hypothesis (H₁):** The mean `avg_txn_value` of high-value records is **significantly greater than** that of low-value records.

**Significance level (α):** 0.05

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy import stats

high_val = df[df['is_high_value'] == 1]['avg_txn_value'].dropna()
low_val  = df[df['is_high_value'] == 0]['avg_txn_value'].dropna()

stat, p_value = stats.mannwhitneyu(high_val, low_val, alternative='greater')
print(f"Mann-Whitney U Statistic : {stat:.4f}")
print(f"P-Value                  : {p_value:.6e}")
print()
if p_value < 0.05:
    print("✅ Result: Reject H₀")
    print("   High-value records have a significantly greater avg_txn_value than low-value records (p < 0.05).")
else:
    print("❌ Result: Fail to Reject H₀")
````

##### Which statistical test have you done to obtain P-Value?

**Mann-Whitney U Test** (one-tailed, alternative = 'greater').

##### Why did you choose the specific statistical test?

The Mann-Whitney U Test was chosen because: (1) `avg_txn_value` is heavily right-skewed (confirmed in Chart 1), violating the normality assumption required by an independent samples t-test; (2) we are comparing two independent groups (high-value vs. low-value records); and (3) it is a robust non-parametric test that works on the rank order of values rather than their raw magnitudes — making it ideal for financial data with extreme outliers.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):** The median `transaction_amount` is **equal** across all transaction types (P2P, Merchant, Recharge, Financial Services, Others).

**Alternate Hypothesis (H₁):** At least one transaction type has a **significantly different** median `transaction_amount` from the others.

**Significance level (α):** 0.05

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
groups = [grp['transaction_amount'].values
          for _, grp in df.groupby('transaction_type')]

stat, p_value = stats.kruskal(*groups)
print(f"Kruskal-Wallis H Statistic : {stat:.4f}")
print(f"P-Value                    : {p_value:.6e}")
print()
if p_value < 0.05:
    print("✅ Result: Reject H₀")
    print("   Transaction amounts differ significantly across transaction types (p < 0.05).")
else:
    print("❌ Result: Fail to Reject H₀")
````

##### Which statistical test have you done to obtain P-Value?

**Kruskal-Wallis H Test** (non-parametric one-way ANOVA).

##### Why did you choose the specific statistical test?

The Kruskal-Wallis test was chosen because: (1) we are comparing **more than two groups** (five transaction types), ruling out Mann-Whitney U; (2) transaction amounts are non-normally distributed (right-skewed), ruling out standard one-way ANOVA which requires normality; (3) Kruskal-Wallis is the appropriate non-parametric extension of ANOVA that tests whether the population medians of all groups are equal, using rank-based statistics robust to outliers and skewness.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):** The mean `registered_users` count is **equal** for high-value and low-value transaction records.

**Alternate Hypothesis (H₁):** High-value records are associated with **significantly higher** `registered_users` counts than low-value records.

##### Why did you choose the specific statistical test?

Same rationale as Hypothesis 1: `registered_users` is heavily right-skewed (a few metro states have extremely large user bases), making the t-test inappropriate. Mann-Whitney U is the correct non-parametric test for comparing two independent groups on a non-normal continuous variable. ## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
print("Missing values before imputation:")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "No missing values ✅")

# Fill any remaining NaN with column median (robust to outliers/skewness)
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

print(f"\nMissing values after imputation: {df.isnull().sum().sum()}")
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

**Technique used: Median Imputation** for all numerical columns.

**Why?** The PhonePe dataset has no raw missing values — nulls only arise from feature engineering steps (e.g., dividing transaction_amount by transaction_count when count is zero produces NaN). For these cases, median imputation is used because: (1) transaction-related features are heavily right-skewed, and the median is resistant to extreme values unlike the mean; (2) it preserves the central tendency of the distribution without introducing bias from high-magnitude outliers; (3) it is simple, fast, and deterministic — ensuring reproducibility.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Using IQR-based Winsorization (capping at 1st and 99th percentile)
def winsorize(series, lower_q=0.01, upper_q=0.99):
    lower = series.quantile(lower_q)
    upper = series.quantile(upper_q)
    return series.clip(lower, upper)

outlier_cols = ['transaction_count', 'avg_txn_value', 'registered_users', 'app_opens', 'users_per_txn']
for col in outlier_cols:
    before_max = df[col].max()
    df[col] = winsorize(df[col])
    after_max = df[col].max()
    print(f"{col:25s} | Before max: {before_max:>15,.0f}  →  After max: {after_max:>15,.0f}")

print("\nOutlier Winsorization complete ✅")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

**Technique used: Winsorization (IQR-based capping at 1st–99th percentile).**

**Why?** (1) Removing outliers entirely would discard valid high-value transaction records — the very records that are most business-critical for fraud detection and premium targeting; (2) Winsorization retains all records while capping extreme values, reducing their disproportionate influence on gradient-based and distance-based models; (3) It is especially important for Logistic Regression which is sensitive to leverage points; tree-based models (Random Forest, XGBoost) are more robust to outliers but still benefit from reduced variance in the input space.

### 3. Categorical Encoding

#### What all categorical encoding techniques have you used & why did you use those techniques?

**Why?** (1) Tree-based models (Random Forest, XGBoost) handle label-encoded integers natively without needing one-hot expansion — keeping feature dimensionality low and avoiding the curse of dimensionality; (2) One-hot encoding `state` would add 35 binary columns, significantly increasing memory and training time with minimal benefit for tree models; (3) For Logistic Regression, the label-encoded features with StandardScaler still produce reasonable results given the regularisation applied. If strict linearity were required, one-hot encoding would be preferred — but for this mixed-model evaluation, Label Encoding is the practical choice.

| Metric | Business Interpretation |
|---|---|
| **Accuracy** | Overall fraction of transaction records correctly classified. High accuracy reduces operational overhead from manual review of misclassified records. |

| **ROC-AUC** | Model's ability to rank high-value above low-value across all decision thresholds. A high AUC means PhonePe can flexibly tune the classification threshold — stricter for fraud (high recall), lenient for marketing (high precision) — without retraining. |

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV)

param_grid_xgb = {
    'n_estimators' : [100, 200],
    'max_depth'    : [3, 5, 7],
    'learning_rate': [0.05, 0.1, 0.2],
    'subsample'    : [0.8, 1.0]
}

gs_xgb = GridSearchCV(
    xgb.XGBClassifier(scale_pos_weight=neg_pos_ratio, eval_metric='auc',
                      use_label_encoder=False, random_state=42, n_jobs=-1),
    param_grid_xgb, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=0
)

# Fit the Algorithm
gs_xgb.fit(X_train, y_train)
print(f"Best Parameters : {gs_xgb.best_params_}")
print(f"Best CV ROC-AUC : {gs_xgb.best_score_:.4f}")

best_xgb = gs_xgb.best_estimator_

# Predict on the model
y_pred_xgb_tuned  = best_xgb.predict(X_test)
y_proba_xgb_tuned = best_xgb.predict_proba(X_test)[:, 1]

print("\n=== XGBoost (Tuned) — Test Set Results ===")
print(f"Accuracy  : {accuracy_score(y_test, y_pred_xgb_tuned):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_xgb_tuned):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred_xgb_tuned):.4f}")
print(f"F1 Score  : {f1_score(y_test, y_pred_xgb_tuned):.4f}")
print(f"ROC-AUC   : {roc_auc_score(y_test, y_proba_xgb_tuned):.4f}")
````

##### Which hyperparameter optimization technique have you used and why?

**GridSearchCV with 5-fold Stratified Cross-Validation**, scoring on ROC-AUC.

**Key parameters tuned:**
- `n_estimators`: Number of boosting rounds — more rounds improve performance until overfitting begins.
- `max_depth`: Shallow trees (3–5) generalise better; deeper trees risk overfitting noisy financial data.
- `learning_rate`: Lower rates (0.05) with more estimators generalise better — classic boosting tradeoff.
- `subsample`: Row subsampling per tree (0.8) adds stochasticity, reducing overfitting similar to dropout in neural networks.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes — tuning `learning_rate` and `max_depth` together typically yields the largest improvement (2–5% in ROC-AUC). Smaller `learning_rate` with more `n_estimators` produces a more robust model that generalises better to unseen data. XGBoost's native handling of class imbalance via `scale_pos_weight` and built-in L1/L2 regularisation (`reg_alpha`, `reg_lambda`) make it the most powerful model in this comparison.

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

For PhonePe's high-value transaction prediction, **ROC-AUC** and **Recall** are the primary metrics considered for positive business impact:

- **ROC-AUC** is the most important metric because it measures the model's discriminative ability across ALL decision thresholds — PhonePe can adjust the threshold based on use case (stricter for fraud detection, more lenient for marketing targeting) without retraining the model. A high AUC guarantees the model is genuinely learning to rank high-value above low-value records.

- **Recall** is critical because missing a true high-value transaction (a False Negative) has real cost — an undetected fraudulent high-value transaction causes direct financial loss, and a missed premium targeting opportunity represents lost revenue. The cost of false negatives outweighs the cost of false positives in this context.

- **F1-Score** is used as the balancing metric when reporting overall performance, since it harmonises precision and recall — giving a single number that penalises models that sacrifice one for the other.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

**Model used: XGBoost (Tuned)**

XGBoost is a gradient boosting algorithm that builds an ensemble of decision trees sequentially. Each new tree corrects the residual errors of the previous ensemble. It uses second-order gradient information (Newton's method) for more accurate leaf value estimates, and regularisation (L1/L2) to prevent overfitting.

**Feature Importance** was assessed using two complementary methods:

**1. XGBoost Built-in Importance (Gain-based):** Measures the average improvement in the loss function brought by each feature across all tree splits. Higher gain = more important feature.

**2. Permutation Importance (Model-agnostic):** Measures the drop in ROC-AUC when a feature's values are randomly shuffled — directly quantifying each feature's contribution to predictive performance on the test set.

Both methods consistently rank **`avg_txn_value`** and **`transaction_count`** as the top two features, followed by **`registered_users`** and **`type_enc`**. This aligns with the Mutual Information analysis in Chart 14 and the visual class separation seen in Chart 12.
